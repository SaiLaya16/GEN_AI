```python
**Objective**

This project builds an AI-powered recruitment screening system that evaluates candidates by analyzing:
Resume content
Job requirements

It automatically:
Extracts skills
Provides explanation for transparency

Why this matters

In real hiring environments:
Recruiters handle large volumes of resumes
Manual screening is slow and inconsistent
Good candidates may get overlooked

This system helps by:

=> Automating evaluation
=> Standardizing candidate scoring
=>Making decisions explainable and fair
=> Reducing recruiter workload

System Workflow:
Resume → Skill Detection → Matching Engine → Score Generator → Explanation Module**
```

```python

from langchain_core.prompts import PromptTemplate

print("Environment ready ✅")
     
```

```python
demo_prompt = PromptTemplate(
    input_variables=["concept"],
    template="Explain {concept} in beginner-friendly language"
)

print(demo_prompt.format(concept="Artificial Intelligence"))
```

```python
job_description = """
Looking for a Data Scientist with skills in Python, Machine Learning, Deep Learning,
NLP, SQL, TensorFlow, and PyTorch.
"""

strong_resume = """
Experienced Data Scientist skilled in Python, Machine Learning, Deep Learning,
NLP, TensorFlow, PyTorch, and SQL. Worked on multiple AI projects.
"""

average_resume = """
Skilled in Python and basic Machine Learning. Some experience with data analysis
and SQL.
"""

weak_resume = """
Basic knowledge of programming. Familiar with MS Excel and simple data entry tasks.
"""
```

```python
def extract_skills(text):
    keywords = [
        "python", "machine learning", "deep learning",
        "nlp", "sql", "tensorflow", "pytorch"
    ]

    text = text.lower()
    found_skills = []

    for skill in keywords:
        if skill in text:
            found_skills.append(skill)

    return found_skills
```

```python
def match_skills(job_skills, resume_skills):
    return list(set(job_skills) & set(resume_skills))
```

```python
def calculate_score(job_skills, matched_skills):
    if len(job_skills) == 0:
        return 0

    score = (len(matched_skills) / len(job_skills)) * 100
    return round(score)
```

```python
def generate_explanation(score, matched, missing):
    return f"""
Final Score: {score}/100

Matched Skills:
{matched}

Missing Skills:
{missing}

The score is calculated by comparing required job skills with candidate skills.
Higher match indicates better suitability.
"""
```

```python
def evaluate_resume(resume, job_description):

    job_skills = extract_skills(job_description)
    resume_skills = extract_skills(resume)

    matched = match_skills(job_skills, resume_skills)
    missing = list(set(job_skills) - set(matched))

    score = calculate_score(job_skills, matched)

    explanation = generate_explanation(score, matched, missing)

    return score, explanation
```

```python
score, explanation = evaluate_resume(strong_resume, job_description)

print("----- Strong Candidate -----")
print(explanation)
```

```python
score, explanation = evaluate_resume(average_resume, job_description)

print("----- Average Candidate -----")
print(explanation)
```

```python
score, explanation = evaluate_resume(weak_resume, job_description)

print("----- Weak Candidate -----")
print(explanation)
```

**Analysis:
Strong Candidate → High skill match and strong job alignment
Average Candidate → Partial match with missing advanced skills
Weak Candidate → Very limited alignment with job requirements**
